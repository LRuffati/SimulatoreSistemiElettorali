```python
import math, random
```

```python
template = """<circle
       style="fill:#{color};stroke:#000000;stroke-width:0;stroke-miterlimit:0;stroke-dasharray:none;stroke-opacity:0"
       id="path4520"
       cx="{x_c}"
       cy="{y_c}"
       r="{r}" />"""
template.format(color="abcdef",x_c=2,y_c=3,r=4)
```

```python
gov=200
opp=160

linee = 0
while gov>(linee**2+linee)*5:
    linee+=1
cols = 5*linee + 5
cols, linee, cols*linee
```

```python
col_a = "00AD90"
count = 0
end=False
for i in range(linee):
    for j in range(cols):
        print(template.format(color=col_a,x_c=20+4*j,y_c=20+4*i,r=2))
        count +=1
        if count == gov:
            end = True
            break
    if end:
        break

col_min = math.ceil(opp/linee)
col_min*linee>opp
col_a = "f28082"
lista_col = ["e0bb00",'6784b9','c61772','66c625']
count = 0
end=False
for i in range(linee):
    for j in range(col_min):
        print(template.format(color=random.choice(lista_col),x_c=20+4*j,y_c=20+4*linee+12+4*i,r=2))
        count +=1
        if count == opp:
            end = True
            break
    if end:
        break
```
